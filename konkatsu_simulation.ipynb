## Initialize

```python
import copy
import random
from candidate_manager import CandidateManager
```

## Run

```python
# random.seed(0)

candidates_max = 100
contraction_rate = 0.5
men_candidate_manager = CandidateManager()
women_candidate_manager = CandidateManager()

men_candidate_manager.initialize_candidates(candidates_max, contraction_rate)
women_candidate_manager.initialize_candidates(candidates_max, contraction_rate)

loop_num = 0
loop_end_flag = False
total_married_pairs_number = 0

while loop_end_flag == False:
    men_name_list = men_candidate_manager.get_name_list()
    women_name_list = women_candidate_manager.get_name_list()

    if loop_num == 0:
        konkatsu_continue_flag = True
    else:
        konkatsu_continue_flag = False

    men_candidate_manager.make_select_new_candidates(women_name_list, konkatsu_continue_flag)
    women_candidate_manager.make_select_new_candidates(men_name_list, konkatsu_continue_flag)

    men_candidates_list = copy.deepcopy(men_candidate_manager.get_candidates_list())
    women_candidates_list = copy.deepcopy(women_candidate_manager.get_candidates_list())

    men_candidate_manager.match_candidates(women_candidates_list)
    women_candidate_manager.match_candidates(men_candidates_list)

    men_candidates_list = copy.deepcopy(men_candidate_manager.get_candidates_list())
    women_candidates_list = copy.deepcopy(women_candidate_manager.get_candidates_list())

    men_married_list = men_candidate_manager.judge_marriage(women_candidates_list)
    women_married_list = women_candidate_manager.judge_marriage(men_candidates_list)

    loop_num += 1
    print("Loop: ", loop_num, " done.")

    total_married_pairs_number += len(men_married_list)
    if len(men_married_list) > 0:
        print(men_married_list)
        print("Married number: ", len(men_married_list))
    if len(women_married_list) > 0:
        print(women_married_list)
        print("Married number: ", len(women_married_list))

    if (men_candidate_manager.check_candidate_konkatsu_all_stopped() == True) and \
       (women_candidate_manager.check_candidate_konkatsu_all_stopped() == True):
        loop_end_flag = True
        print("Konkatsu stopped.")

print("Loop Number: ", loop_num)
print("Total Married Pairs Number: ", total_married_pairs_number)
```
